```python
# Clean table
def clean(df):
    
    # Clean
    df[1] = df[1].str[:5]
    df[['pilot_nationality','pilot']] = df[2].str.split("\n",expand=True)
    df[['country','launch']] = df[3].str.split("\n",expand=True)
    df[5] = df[5].str.rstrip('km')
    df[6] = df[6].str.rstrip('p.')
    
    # Rename and drop    
    df.rename(columns={1: 'hour', 5: 'distance', 6: 'points', 7: 'km/h', 11: 'date'}, inplace=True)
    df.drop([0,2,3,4,8,9,10], axis=1, inplace=True)
    df = df.iloc[:,[4,0,5,6,7,8,1,2,3]]
    
    # Clean formats
    df[['distance', 'points', 'km/h']] = df[['distance', 'points', 'km/h']].apply(pd.to_numeric, axis=1)
    df['date'] = pd.to_datetime(df['date'])
    df['hour'] = pd.to_datetime(df['hour']).dt.strftime('%H:%M:%S')
    
    return df
```
